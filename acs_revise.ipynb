```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from neuralforecast import NeuralForecast
from neuralforecast.losses.pytorch import MQLoss, DistributionLoss
from neuralforecast.models import LSTM
from pylab import mpl

# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

```python
init_data=pd.read_excel('SSA_nanbao.xlsx')
# unique_id= pd.DataFrame(np.arange(0, 2900, 1),columns=['unique_id'])
init_data.insert(loc=0,column='unique_id',value='1')
init_data.columns=['unique_id','ds','y']
train_data=init_data[:2320]
test_data=init_data[2320:]
init_data
```

```python
fcst = NeuralForecast(
    models=[LSTM(h=10, input_size=-1,
                 loss=DistributionLoss(distribution='Normal', level=[80, 90]),
                 scaler_type='robust',
                 encoder_n_layers=2,
                 encoder_hidden_size=128,
                 context_size=10,
                 decoder_hidden_size=128,
                 decoder_layers=2,
                 max_steps=200,
                 )
    ],
    freq='M'
)
fcst.fit(df=train_data)
```

```python
from neuralforecast.models import TCN
from neuralforecast.losses.pytorch import GMM, MQLoss, DistributionLoss
fcst = NeuralForecast(
    models=[TCN(h=10,
                input_size=-1,
                #loss=DistributionLoss(distribution='Normal', level=[80, 90]),
                loss=GMM(n_components=7, return_params=True, level=[80,90]),
                learning_rate=5e-4,
                kernel_size=2,
                dilations=[1,2,4,8,16],
                encoder_hidden_size=128,
                context_size=10,
                decoder_hidden_size=128,
                decoder_layers=2,
                max_steps=200,
                scaler_type='robust',
                )
    ],
    freq='M'
)
fcst.fit(df=train_data)
```

```python
from neuralforecast.models import GRU
from neuralforecast.losses.pytorch import MQLoss, DistributionLoss
fcst = NeuralForecast(
    models=[GRU(h=10,input_size=-1,
                loss=DistributionLoss(distribution='Normal', level=[80, 90]),
                scaler_type='robust',
                encoder_n_layers=2,
                encoder_hidden_size=128,
                context_size=10,
                decoder_hidden_size=128,
                decoder_layers=2,
                max_steps=200,
                )
    ],
    freq='M'
)
fcst.fit(df=train_data)
```

```python
pred_test=fcst.predict(df=train_data)
```

```python
pred_test_TCN=fcst.predict(df=train_data)
```

```python
pred_test_GRU=fcst.predict(df=train_data)
```
